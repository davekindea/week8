```python
import pandas as pd
data_fraud=pd.read_csv("/content/cleaned_fruad_data.csv")
```

```python
data_creditcard=pd.read_csv("/content/creditcard.csv")
```

```python
for col in data_fraud.columns:
    mode_value = data_fraud[col].mode()[0]
    data_fraud[col].fillna(mode_value, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report

# Separate Features and Target Variables
X = data_fraud.drop(columns='class')
y = data_fraud['class']

# Create Train & Test Data
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,
                                                	stratify =y,
                                                	random_state = 13)

# Build the model
rf_fraud = RandomForestClassifier(max_features=2, n_estimators =100 ,bootstrap = True)

rf_fraud.fit(X_train, y_train)

# Make prediction on the testing data
y_pred = rf_fraud.predict(X_test)

# Classification Report
print(classification_report(y_pred, y_test))
```

```python
# Create Decision Tree classifer object
from sklearn.tree import DecisionTreeClassifier
clf_fraud = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf_fraud = clf_fraud.fit(X_train,y_train)

#Predict the response for test dataset
y_pred_fraud = clf_fraud.predict(X_test)
print(classification_report(y_pred_fraud, y_test))
```

```python
data_fraud.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg_fraud = LogisticRegression()

# fit the model with data
logreg_fraud.fit(X_train, y_train)

y_pred_fraud = logreg_fraud.predict(X_test)
print(classification_report(y_pred_fraud, y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_fraud = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model_fraud.fit(X_train, y_train)
y_pred = model_fraud.predict(X_test)
print(classification_report(y_pred, y_test))
```
